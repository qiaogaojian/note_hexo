# 高等数学基础

## 三角函数

### 1.1 三角函数的概念

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,1000)
a = np.sin(x)
b = np.cos(x)
c = np.tan(x)
d = 1/np.tan(x)

plt.figure(figsize=(8,4))
plt.plot(x,a,label='$sin(x)$',color='green',linewidth=0.5)

plt.xlim(-np.pi0,np.pi)
plt.ylim(-1,1)


ax=plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))  
ax.spines['left'].set_position(('axes',0.5)) 


plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,b,label='$cos(x)$',color='red',linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)
plt.legend()
plt.show()
```

```python
x = np.linspace(-10,10,1000)
a = np.sin(x)
b = np.cos(x)
c = np.tan(x)
d = 1/np.tan(x)

plt.figure(figsize=(8,4))
plt.plot(x,a,label='$sin(x)$',color='green',linewidth=0.5)
plt.plot(x,b,label='$cos(x)$',color='red',linewidth=0.5)
plt.plot(x,c,label='$tan(x)$',color='blue',linewidth=0.5)
plt.plot(x,d,label='$cot(x)$',color='purple',linewidth=0.5)

plt.xlabel('Time(s)')
plt.ylabel('Vlot')
plt.xlim(0,10)
plt.ylim(-3,3)
plt.legend()
plt.show()
```
